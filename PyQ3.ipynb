```python
# Define the department to code mapping
department_mapping = {
'HR': 'HR001',
'Sales': 'SAL002',
'Manager': 'MGR003',
'Developer': 'DEV004'
}
```

```python
# Convert the dictionary to a DataFrame
department_df = pd.DataFrame(list(department_mapping.items()), columns=['Department', 'Department_Code'])
```

```python
# Display the DataFrame
print(department_df)
```

```python
# Q1) Create Age_Bucketed column
bins = [0, 20, 30, 40, 50, 120]
labels = ["<20 years", "20-30 years", "30-40 years", "40-50 years", ">50 years"]
df['Age_Bucketed'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df
```

```python
# Q2) Create Function_Code column using department_df mapping
df['Function_Code'] = df['Function'].map(department_mapping)
```

```python
df
```

```python
# Q3) Create Outlier column using IQR method for Salary
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['Outlier'] = np.where((df['Salary'] < lower_bound) | (df['Salary'] > upper_bound), 
                        "Outlier", "Not an Outlier")
```
