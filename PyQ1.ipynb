```python
import pandas as pd
```

```python
# Load the Excel file
file_path = "C:\\Users\\Channabasava\\Desktop\\Test\\PyQ1.xlsx" 
xls = pd.ExcelFile(file_path)
```

```python
# Read the required sheets into dataframes
df_sales = pd.read_excel(xls, sheet_name="Sales")
df_tv_grps = pd.read_excel(xls, sheet_name="TV GRPs")
df_video_impressions = pd.read_excel(xls, sheet_name="Video Impressions")
```

```python
# Merge all three DataFrames
merged_df  = df_sales.merge(df_tv_grps, on='Timestamp', how='left')\
                   .merge(df_video_impressions, on='Timestamp', how='left')
```

```python
merged_df
```

```python
# Save to the Output sheet
with pd.ExcelWriter(file_path, mode="a", if_sheet_exists="replace") as writer:
    merged_df.to_excel(writer, sheet_name="Output", index=False)
```
